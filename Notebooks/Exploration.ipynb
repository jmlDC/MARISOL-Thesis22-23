```python
from sklearn.base import BaseEstimator, TransformerMixin
class GetDummiesTransformer(BaseEstimator, TransformerMixin):
    def __init__(self, *args, pandas_params={}, **kwargs):
        super().__init__(*args, **kwargs)
        self._pandas_params = pandas_params
    def fit(self, X, y=None):
        return self
    def transform(self, X, y=None):
        return pd.get_dummies(X, **self._pandas_params)

bert_ohe = GetDummiesTransformer(pandas_params={'columns':['BERT-code_frames']}).transform(bert)
```

### CHI-square

```python
import scipy.stats as stats
import numpy as np

for x in range(15):
    observed_data = dataframe.query(f'Frame == {x+1}')['Observed Count'].tolist()
    expected_data = dataframe.query(f'Frame == {x+1}')['Expected Count'].tolist()
    chi_square_test_statistic1 = 0
        # Chi-Square Goodness of Fit Test
    chi_square_test_statistic, p_value = stats.chisquare(
        observed_data, expected_data)
    print(f"Frame {x+1}")
    # chi square test statistic and p value
    print('chi_square_test_statistic is : ' +
        str(chi_square_test_statistic))
    print('p_value : ' + str(p_value))

    # find Chi-Square critical value
    print("crit val:", stats.chi2.ppf(1-0.05, df=14))
    print("")
```
