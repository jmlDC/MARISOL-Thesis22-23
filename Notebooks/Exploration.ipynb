```python
gmadf, pnadf, rapplerdf, sunstartdf, mbdf = pd.read_csv(f'{dir}predictions/SBP/gmaBERT-last.csv'), pd.read_csv(f'{dir}predictions/SBP/pnaBERT-last.csv'), pd.read_csv(f'{dir}predictions/SBP/rapplerBERT-last.csv'),pd.read_csv(f'{dir}predictions/SBP/sunstarBERT-last.csv'),pd.read_csv(f'{dir}predictions/SBP/mbBERT-last.csv')


for df in [gmadf, pnadf, rapplerdf, sunstartdf, mbdf]:
    for x in  df.columns:
        if (x!= 'Text'):
            df[x] = df[x].astype(int)

```

```python
fig, axes = plt.subplots(2, 3, figsize=(18, 10))

# fig.suptitle('Code Frames for each Model')
gma = sns.countplot(ax=axes[0,0], data=gmadf, x="Annotate",palette="husl")
pna = sns.countplot(ax=axes[0,1], data=pnadf, x="Annotate",palette="husl")
rappler = sns.countplot(ax=axes[0,2],data=rapplerdf, x="Annotate" ,palette="husl")
sunstar = sns.countplot(ax=axes[1,0], data=sunstartdf, x="Annotate" ,palette="husl")
mb = sns.countplot(ax=axes[1,1], data=mbdf, x="Annotate" , hue='Annotate', dodge=False ,palette="husl")


gma.set_xlabel('GMA BERT code_frames')
gma.set_ylabel('count of frame instances')

pna.set_xlabel('Philippine News Agency BERT code_frames')
pna.set_ylabel('count of frame instances')

rappler.set_xlabel('Rappler BERT code_frames')
rappler.set_ylabel('count of frame instances')

sunstar.set_xlabel('Sunstar BERT code_frames')
sunstar.set_ylabel('count of frame instances')

mb.set_xlabel('Manila Bulletin BERT code_frames')
mb.set_ylabel('count of frame instances')


fig.delaxes(axes[1][2])

# axes[1,1].legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)

axes[1,1].legend(labels = ['1: Economic', '2: Capacity and Resources', '3: Morality', '4: Fairness and Equality', '5: Constitutionality and jurisprudence',
                     '6: Policy Prescription and Evaluation', '7: Law and Order, Crime and Justice', '8: Security and Defense', '9: Health and Safety', '10: Quality of Life',
                     '11: Cultural Identity', '12: Public Opinion', '13: Political', '14: External Regulation and Reputation', '15: Other'
                     ], loc='upper right', bbox_to_anchor=(2, 1))
plt.show()
```
