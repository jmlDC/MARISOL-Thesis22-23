## Group A

```python
a1 = pd.read_excel(f"{dir_actual}1A-groupA-annotation.xlsx", usecols=["Text","Annotate"])
a1.rename(columns = {'Annotate':'A1-Annotations'}, inplace = True)
a1
```

```python
a2 = pd.read_excel(f"{dir_actual}2A-groupA-annotation.xlsx", usecols=["Text","Annotate"])
a2.rename(columns = {'Annotate':'A2-Annotations'}, inplace = True)
a2
```

```python
a3 = pd.read_excel(f"{dir_actual}3A-groupA-annotation.xlsx", usecols=["Text","Annotate"])
a3.rename(columns = {'Annotate':'A3-Annotations'}, inplace = True)
a3
```

```python
concated_a = pd.concat([a1,a2,a3], axis=1,keys = ['a1', 'a2',"a3"]) #[['Text',"A1-Annotations", "A2-Annotations", "A3-Annotations"]]
# concated_a['Text'] = a1.t
concated_a
```

```python
concatenated_a = concated_a.T.drop_duplicates().T
```

```python
df = concatenated_a
df
```

```python
import numpy as np
df['diff']=np.where(df['a1']['Text']==df['a2']['Text'],
                    True,
                    False)
df
```

```python
for i, r in df.iterrows():
    if r['diff'].bool()==False:
        print(i)
```
