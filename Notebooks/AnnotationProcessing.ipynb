# Majority Ruling

```python
# https://stackoverflow.com/a/43990682
def getMajority(df, a,b,c):
    mostCommonVote=[]
    for row in df[[a,b,c]].values:
        votes, values = np.unique(row, return_counts=True)
        if np.all(values<=1):
                mostCommonVote.append(False)
        else:
            mostCommonVote.append( votes[np.argmax(values)] )

    df['Majority-Frame'] = mostCommonVote
```

### 2nd Fleiss kappa

```python
# https://medium.com/@nick.acosta/an-introduction-to-fleiss-kappa-36d3a9e7d606
def fleiss_kappa(lists, classes):
    n = len(lists)
    N = len(lists[0])
    k = len(classes)

    nij = []
    for i in range(N):
        nij.append([0]*k)


    for i in range(len(lists)):
        for j in range(len(lists[i])):
            nij[j][classes.index(lists[i][j])] += 1

    P = []
    for i in nij:
        P.append(1/(n*(n-1))*(sum([j*j for j in i])-n))

    return (((sum(P)/N)-(sum([y*y for y in [x/(N*n) for x in[sum(i) for i in zip(*nij)]]])))/(1-sum([y*y for y in [x/(N*n) for x in[sum(i) for i in zip(*nij)]]]))+1)/2
```

```python
classes = ['-', 'o', '+'] # negative, neutral, positive

rater1 = ['+', '+', '+', '+', 'o', '+', 'o', '-', '+', 'o']
rater2 = ['+', '+', 'o', 'o', 'o', '+', 'o', 'o', '+', 'o']
rater3 = ['+', '+', '+', '+', 'o', '+', 'o', '-', '+', '+']

rate = [rater1, rater2, rater3]

print(fleiss_kappa(rate, classes))
```

### 3rd Fleiss kappa

```python
ratings = [(1, 'yes')] * 10 + [(1, 'no')] * 0  + \
[(2, 'yes')] * 8  + [(2, 'no')] * 2  + \
[(3, 'yes')] * 9  + [(3, 'no')] * 1  + \
[(4, 'yes')] * 0  + [(4, 'no')] * 10 + \
[(5, 'yes')] * 7  + [(5, 'no')] * 3

ratings
```

```python
# https://github.com/amirziai/learning/blob/master/statistics/Inter-rater%20agreement%20kappas.ipynb
# Adapted from https://gist.github.com/ShinNoNoir/4749548
def fleiss_kappa_2(ratings, n):
    '''
    Computes the Fleiss' kappa measure for assessing the reliability of
    agreement between a fixed number n of raters when assigning categorical
    ratings to a number of items.

    Args:
        ratings: a list of (item, category)-ratings
        n: number of rows raters annotated
    Returns:
        the Fleiss' kappa score

    See also:
        http://en.wikipedia.org/wiki/Fleiss'_kappa
    '''
    items = set()
    categories = set()
    n_ij = {}

    for i, c in ratings:
        items.add(i)
        categories.add(c)
        n_ij[(i,c)] = n_ij.get((i,c), 0) + 1

    N = len(items)

    p_j = dict(((c, sum(n_ij.get((i, c), 0) for i in items) / (1.0 * n * N)) for c in categories))
    P_i = dict(((i, (sum(n_ij.get((i, c), 0) ** 2 for c in categories) - n) / (n * (n - 1.0))) for i in items))

    P_bar = sum(P_i.values()) / (1.0 * N)
    P_e_bar = sum(value ** 2 for value in p_j.values())

    kappa = (P_bar - P_e_bar) / (1 - P_e_bar)

    return kappa
```
