# Majority Ruling

```python
# https://stackoverflow.com/a/43990682
def getMajority(df, a,b,c):
    mostCommonVote=[]
    for row in df[[a,b,c]].values:
        votes, values = np.unique(row, return_counts=True)
        if np.all(values<=1):
                mostCommonVote.append(False)
        else:
            mostCommonVote.append( votes[np.argmax(values)] )

    df['Majority-Frame'] = mostCommonVote
```

### 2nd Fleiss kappa

```python
# https://medium.com/@nick.acosta/an-introduction-to-fleiss-kappa-36d3a9e7d606
def fleiss_kappa(lists, classes):
    n = len(lists)
    N = len(lists[0])
    k = len(classes)

    nij = []
    for i in range(N):
        nij.append([0]*k)


    for i in range(len(lists)):
        for j in range(len(lists[i])):
            nij[j][classes.index(lists[i][j])] += 1

    P = []
    for i in nij:
        P.append(1/(n*(n-1))*(sum([j*j for j in i])-n))

    return (((sum(P)/N)-(sum([y*y for y in [x/(N*n) for x in[sum(i) for i in zip(*nij)]]])))/(1-sum([y*y for y in [x/(N*n) for x in[sum(i) for i in zip(*nij)]]]))+1)/2
```

```python
classes = ['-', 'o', '+'] # negative, neutral, positive

rater1 = ['+', '+', '+', '+', 'o', '+', 'o', '-', '+', 'o']
rater2 = ['+', '+', 'o', 'o', 'o', '+', 'o', 'o', '+', 'o']
rater3 = ['+', '+', '+', '+', 'o', '+', 'o', '-', '+', '+']

rate = [rater1, rater2, rater3]

print(fleiss_kappa(rate, classes))
```
