## finding diff

```python
# https://kanoki.org/2022/08/01/pandas-compare-columns-in-two-dataframes/
def fdiff(df,k1,k2):
    df['diff']=np.where(df[k1]['Text']==df[k2]['Text'], True, False)
    for i, r in df.iterrows():
        if r['diff'].bool()==False:
            print(i)
```

# Majority Ruling

```python
# https://stackoverflow.com/a/43990682
def getMajority(df, a,b,c):
    mostCommonVote=[]
    for row in df[[a,b,c]].values:
        votes, values = np.unique(row, return_counts=True)
        if np.all(values<=1):
                mostCommonVote.append(False)
        else:
            mostCommonVote.append( votes[np.argmax(values)] )

    df['Majority-Frame'] = mostCommonVote
```

```python
getMajority(concatenated_c,"C1-Annotations", "C2-Annotations", "C3-Annotations")
concatenated_c
```
